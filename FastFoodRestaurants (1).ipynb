4.2. Missing Data and its imputation

```python
#missing data
total = fr.isnull().sum().sort_values(ascending=False)
percent = (fr.isnull().sum()/fr.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
print(missing_data)
```

```python
#dealing with missing data
#Letâ€™s get rid of the variable â€œwebsitesâ€, just to make our lives easier.
fr = fr.drop(['websites'], axis= 1)
fr.head(5)
```
